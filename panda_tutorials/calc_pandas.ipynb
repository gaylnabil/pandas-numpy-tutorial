### 5- Updating Rows and Columns - Modifying Data Within DataFrames

```python
data_frames.columns
```

```python
data_frames.columns = [col.replace('_', '/').upper() for col in data_frames.columns]
data_frames.columns
```

```python
data_frames.rename(columns={'CITY': 'City'}, inplace=True)
data_frames.columns
```

```python
# Rename and replace all columns
data_frames.columns = [name.replace('/', '_').lower() for name in data_frames.columns]
data_frames.columns
```

```python
# Reset indexing
data_frames.reset_index(inplace=True)
```

```python
# Edit a specific cell in the table
data_frames.loc[3, ['city', 'city_ascii', 'state_id']] = ['Manitoba', 'Manitoba', 'MB']
data_frames.loc[[1,2,3], ['city', 'city_ascii', 'state_id']] 
```

```python
# Find and Edit element in the row
# Manhattan	Manhattan	NY
data_filter = (data_frames['city'] == 'Manitoba')
data_frames.loc[data_filter, ['city', 'city_ascii', 'state_id']] = ['Manhattan', 'Manhattan', 'NY']
data_frames.loc[:5, ['city', 'city_ascii', 'state_id']] 
```

```python
# Edit an element in all rows of the table
data_frames['city'] = data_frames['city'].str.upper()
data_frames.loc[:5, ['city', 'city_ascii', 'state_id']] 
```

```python
# Using apply function: .apply(function)
data_frames['city'].apply(len)
```

```python
data_frames['city'] = data_frames['city'].apply(lambda x: x.capitalize())
data_frames.loc[:5, ['city', 'city_ascii', 'state_id']] 
```

```python
filter_data = (data_frames['incorporated'] == True)
data_frames.loc[filter_data, 'population']
```

```python
average = int(data_frames.loc[filter_data, 'population'].sum() / data_frames.loc[filter_data, 'population'].count())
print('average:', average)
```

```python
# Drop specific columns from the DataFrame
data_frames = data_frames.drop(columns={'zips', 'source', 'ranking'})
data_frames
```

```python
# Add a new Column 'full_state' with Concatenation of columns 'state_id' and 'state_name'
data_frames['full_state'] = data_frames['state_id'] + ' - ' + data_frames['state_name']
data_frames
```

```python
# Drop specific columns from the DataFrame
if data_frames.columns.isin(['state_id', 'state_name']).any():
    data_frames.drop(columns={'state_id', 'state_name'}, inplace=True)
data_frames
```
