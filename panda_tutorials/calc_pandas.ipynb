### 5- Updating Rows and Columns - Modifying Data Within DataFrames

### 6- Add/Remove Rows and Columns From DataFrames

### 8- Grouping and Aggregating - Analyzing and Exploring Your Data

```python
# median(): Having a value in the middle of series of values
# median value for one column of series
data_frame['population'].median()
```

```python
# describe(): Having all values of (count, mean, std, min, 25%, median = 50%, 75%, max)
data_frame.describe()
```

```python
# Count how many times of each value in the column
data_frame['county_name'].value_counts()
```

```python
# Count how many times of each value in the column
# normalize = True: means the percentage of values
data_frame['county_name'].value_counts(normalize = True)
```

```python
# Using GroupBy in Pandas
# Group by 'county_name' with value 'Franklin'
county_grp = data_frame.groupby(data_frame['county_name'], group_keys=True)
# county_name_grp = data_frame.groupby(data_frame['county_name'])
county_grp.get_group('Franklin')

# the same results with filter 
# filter_data = (data_frame['county_name'] == 'Franklin')
# data_frame[filter_data]
# data_frame[filter_data]['state_name'].value_counts()
```

```python
# Group by 'county_name' with value 'Franklin' with the largest value 'population'
county_grp.get_group('Franklin').nlargest(1, 'population')
```

```python
# Return value counts of state_name grouped by 'county_name'
county_grp['state_name'].value_counts()
```

```python
# Return value counts of state_name grouped by 'county_name' and filtered by 'county_name'-> .loc['Yuma'] means county_name == 'Yuma'
county_grp['state_name'].value_counts(normalize=True).loc['Yuma'] * 100 
```

```python
county_grp['population'].median()
```

```python
county_grp['population'].median().loc['Acadia']
```

```python
county_grp['population'].describe()
```

```python
county_grp['population'].aggregate(['median', 'mean', 'min', 'max'])
```

```python
sum_virginia_state = county_grp['state_name'].apply(lambda x: x.str.contains('Virginia').sum())
sum_virginia_state
```

```python
state_respondents = data_frame['county_name'].value_counts()
state_respondents
```

```python
virginia_state_df = pd.concat([state_respondents, sum_virginia_state], axis='columns', sort=True)
```

```python
virginia_state_df.rename(columns={'county_name': 'number_country', 'state_name': 'number_state'}, inplace=True)
virginia_state_df.sort_index()
```

```python
virginia_state_df['percent']=virginia_state_df[['number_country', 'number_state']].apply(
    lambda x: 0 if x['number_country']== 0 else ((x['number_state'] / x['number_country']) * 100).round(2), axis=1
    )
virginia_state_df
```

```python
virginia_state_df.sort_values(by=['number_country'], ascending=False).head(5)
```
