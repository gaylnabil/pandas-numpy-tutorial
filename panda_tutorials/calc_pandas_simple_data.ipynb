### 4- Filtering - Using Conditionals to Filter Rows and Columns

```python
pd_filter = ((df['first_name'] == "Nabil") | (df['last_name'] == "GAYL"))
```

```python
df[pd_filter]
```

```python
# The same filter results of 'df[fn_filter]'
df.loc[pd_filter]
```

```python
# get filter with specified column name and the opposite of 'fn_filter'
df.loc[~pd_filter, ['first_name', 'email']]
```

### 5- Updating Rows and Columns - Modifying Data Within DataFrames

```python
df.columns
```

```python
df.columns = [name.replace('_', ' ').upper() for name in df.columns]
df.columns
```

```python
# Rename a specific column
df.rename(columns={'EMAIL': 'Email'}, inplace=True)
df.columns
```

```python
# Rename and replace all columns
df.columns = [name.replace(' ', '_').lower() for name in df.columns]
df.columns
```

```python
# Edit a specific cell in the table
df.iloc[4] =  ['kamal.sabiri@yahoo.com', 'Kamal', 'SABIRI', 29]
df
```

```python
# Edit a specific row in the table with specified columns
df.loc[4, ['email', 'first_name']] =  ['ahmed.sabiri@yahoo.com', 'Ahmed']
df
```

```python
# Find and Edit element in the row
data_filter = (df['email'] == 'Mourabiti.NABIL@outlook.com')
df.loc[data_filter, 'email'] = df.loc[data_filter, 'email'].str.lower()
df
```

```python
# Edit an element in all rows of the table
df['email'] = df['email'].str.upper()
df
```

```python
# Using apply function: .apply(function)
df['email'].apply(len)
```

```python
def email_lower(email):
    return email.lower()
```

```python
df['email'] = df['email'].apply(email_lower)
df
```

```python
# Using applymap function: .applymap(function), return a table
df[['email', 'first_name', 'last_name']].applymap(str.upper)
df
# df.iloc[:, :-1].applymap(str.upper)
```
